```python
import pandas as pd
import yfinance as yf
import statsmodels.api as sm
```

```python
data=yf.download("^GSPC",start="2001-01-03",end="2005-12-31")
```

```python
df =data["Adj Close"].pct_change()*100
```

```python
df
```

```python
df=df.rename("Today")
```

```python
df
```

```python
df=df.reset_index()
```

```python
df
```

```python
for i in range(1,6):
    df["Lag"+str(i)]=df["Today"].shift(i)
```

```python
df
```

```python
df["Volume"]=data.Volume.shift(1).values/1000_000_000
```

```python
df=df.dropna()
```

```python
df
```

```python
df["Direction"]=[1 if i > 0 else 0 for i in df["Today"]]
```

```python
df
```

```python
df=sm.add_constant(df)
```

```python
df
```

```python
print(df.columns)
df_reset = df.reset_index(drop=True)
print(df_reset.columns)
```

```python
print(df.head())
```

```python
X = df_reset[['const','Lag1', 'Lag2', 'Lag3', 'Lag4', 'Lag5', 'Volume']]
```

```python
y=df.Direction
```

```python
import statsmodels.api as sm

# Reset indices of X and y to ensure alignment
X_reset = X.reset_index(drop=True)
y_reset = y.reset_index(drop=True)

# Create the logistic regression model
model = sm.Logit(y_reset, X_reset)

# Fit the model
result = model.fit()

# Print summary of the model
print(result.summary())
```

```python
prediction=result.predict(X)
```

```python
prediction
```

```python
y
```

```python
def confusion_matrix(act,pred):
    predtrans=["Up" if i > 0.5 else "Down" for i in pred]
    actuals=["Up" if i > 0 else "Down" for i in act]
    confusion_matrix=pd.crosstab(pd.Series(actuals),
                                 pd.Series(predtrans),
                                 rownames=["Actual"],
                                 colnames=["Predicted"])
    
    return confusion_matrix
```

```python
confusion_matrix(y,prediction)
```
