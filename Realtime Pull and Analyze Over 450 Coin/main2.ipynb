```python
def getreturn(coin):
    coin_frame=df[df.coin==coin]
    coin_frame=coin_frame.sort_values("ts")
    coin_frame=coin_frame.set_index("ts")
    return coin_frame.price[-1]/coin_frame.price[0]-1
```

```python
getreturn("BTCUSDT")
```

```python
rets=[]

df=pd.read_csv("coinprices.csv",header=None)
df.columns=["ts","price","coin"]

for coin in df.coin.unique():
    rets.append(getreturn(coin))
```

```python
rets
```

```python
series_=pd.Series(rets,index=df.coin.unique())
```

```python
series_.nlargest(5)
```
